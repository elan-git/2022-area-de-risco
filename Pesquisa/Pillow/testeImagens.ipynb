##  Manipulação de novas imagens e/ou contrução delas

#### Comandos básicos ainda não testados, mas importantes: 

image.save(): comando para salvar imagens;

image.crop((left, top, rigth, bottom)): comando para cortar uma imagem com os parâmetros relacionados ao tamanho de cada direção (esquerda, cima, direita, baixo). Importante prestar atenção que left e top formam o ponto inicial do corte e rigth bottom formam o ponto final.

image.rotate(angulo): rotação de imagem recebendo o ângulo;

#### Copiar imagens

```python
cop_im = im.copy()
plt.imshow(cop_im)
```

#### Transposing



```python
#Comandos
ori_im1 = im.transpose(Image.Transpose.FLIP_LEFT_RIGHT)
ori_im2 = im.transpose(Image.Transpose.FLIP_TOP_BOTTOM)
ori_im3 = im.transpose(Image.Transpose.ROTATE_90)
ori_im4 = im.transpose(Image.Transpose.ROTATE_180)
ori_im5 = im.transpose(Image.Transpose.ROTATE_270)
ori_im6 = im.transpose(Image.Transpose.TRANSPOSE)
ori_teste = im.transpose(Image.Transpose.TRANSVERSE)

#Seção para organização dos plots das imagens
plt.figure(figsize=(10, 10))
plt.subplot(3,2,1)
plt.imshow(ori_im1)
plt.title("Direita para esquerda")
plt.subplot(3,2,2)
plt.imshow(ori_im2)
plt.title("Cima para baixo")
plt.subplot(3,2,3)
plt.imshow(ori_im3)
plt.title("Rotação de 90°")
plt.subplot(3,2,4)
plt.imshow(ori_im4)
plt.title("Rotação 180°")
plt.subplot(3,2,5)
plt.imshow(ori_im5)
plt.title("Rotação 270°")
plt.subplot(3,2,6)
plt.imshow(ori_im6)
plt.title("Transpose")
```

O transpose atua em transformar as linhas em colunas e as colunas em linhas.

```python
plt.imshow(ori_teste)
plt.title(".")
```

#### Definir um novo tamanho de imagem

```python
newsize = (800, 400)

#Comandos
res_im1 = im.resize(newsize, Image.Resampling.BILINEAR)
res_im2 = im.resize(newsize, Image.Resampling.NEAREST)
res_im3 = im.resize(newsize, Image.Resampling.BOX)
res_im4 = im.resize(newsize, Image.Resampling.HAMMING)
res_im5 = im.resize(newsize, Image.Resampling.BICUBIC)
res_im6 = im.resize(newsize, Image.Resampling.LANCZOS)

#Seção para organização dos plots das imagens
plt.figure(figsize=(10, 10))
plt.subplot(3,2,1)
plt.imshow(res_im1)
plt.title("Bilinear")
plt.subplot(3,2,2)
plt.imshow(res_im2)
plt.title("Nearest")
plt.subplot(3,2,3)
plt.imshow(res_im3)
plt.title("Box")
plt.subplot(3,2,4)
plt.imshow(res_im4)
plt.title("Hamming")
plt.subplot(3,2,5)
plt.imshow(res_im5)
plt.title("Bicubic")
plt.subplot(3,2,6)
plt.imshow(res_im6)
plt.title("Lanczos")
```
